```python
import plotly.graph_objects as go

import pandas as pd

# Load data
df = pd.read_csv(
    "https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]

# Create figure
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=list(df.Date), y=list(df.High)))

# Set title
fig.update_layout(
    title_text="Time series with range slider and selectors"
)

# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

frames        = np.arange(0, 600, 1/30) #time in frames
fluoresence   = np.sin(frames)

data = {"frames": frames, "fluoresence": fluoresence}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Create figure
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=list(df.frames), y=list(df.fluoresence)))

# Set title
fig.update_layout(
    title_text="Time series with range slider and selectors"
)
fps = 30
# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=fps*1,
                     label="1s",
                     step="second",
                     stepmode="backward"),
                dict(count=fps*5,
                     label="5s",
                     step="second",
                     stepmode="backward"),
                dict(count=fps*30,
                     label="30s",
                     step="second",
                     stepmode="backward"),
                dict(count=fps*60,
                     label="1m",
                     step="second",
                     stepmode="backward"),
                dict(count=fps*300,
                     label="5m",
                     step="second",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="time"
    )
)

fig.show()
```

```python
import plotly.graph_objects as go # or plotly.express as px
fig = go.Figure() # or any Plotly Express function e.g. px.bar(...)
# fig.add_trace( ... )
# fig.update_layout( ... )

import dash
import dash_core_components as dcc
import dash_html_components as html
from dash import Dash, html, dcc, Input, Output, callback

app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=fig),
    dcc.RangeSlider(0, 20, 1, value=[5, 15], id='time-range'),
    html.Div(id='output-container-range-slider')
])

@callback(
    Output('output-container-range-slider', 'children'),
    Input('time-range', 'value'))
def update_output(value):
    return 'You have selected "{}"'.format(value)

app.run_server(debug=True, use_reloader=False)  # Turn off reloader if inside Jupyter
```
